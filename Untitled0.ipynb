```python
# Design Matrix

def dmat_prep(df):
  if isinstance(df, pd.DataFrame):
    dmat = df.drop('Ancestry', axis = 1)
  else:
    dmat = np.delete(df, 10, 1)
    dmat = pd.DataFrame(dmat, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
    dmat = dmat.drop('Ancestry', axis = 1)



  
  for x in ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10']:
    dmat[x] = dmat[x].transform(lambda x: (x-x.mean())/x.std())

  dmat = dmat.to_numpy()
  ones = np.ones(np.shape(df)[0])
  dmat = np.hstack((ones[:, np.newaxis], dmat))

  return(dmat)
```

```python
# Response Matrix
# African, European, East Asian, Oceanian, or Native American

def rmat_prep(df):
  if isinstance(df, np.ndarray):
    df = pd.DataFrame(df, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
  rmat = []
  for x in df['Ancestry']:
    if x == 'African':
      rmat.append([1, 0, 0, 0, 0])
    if x == 'European':
      rmat.append([0, 1, 0, 0, 0])
    if x == 'EastAsian':
      rmat.append([0, 0, 1, 0, 0])
    if x == 'Oceanian':
      rmat.append([0, 0, 0, 1, 0])
    if x == 'NativeAmerican':
      rmat.append([0, 0, 0, 0, 1])
    
  rmat = np.array(rmat)
  return(rmat)
```

```python
# Gradient Decent 
def gradD(X, Y, lam=[10000]):
  # Parameter Matrix initialized to all 0s
  # df = df_train
  if isinstance(X, pd.DataFrame):
    X = dmat_prep(X)
  if isinstance(Y, pd.DataFrame):
    Y = rmat_prep(Y)
  a = 0.00001
  # l = 10000

  Blist = []
  for l in lam:
    B = np.zeros((11, 5))
    # Probability Matrix initialized to all 0s
    for x in range(10000):
      Z = np.zeros((11, 5))
      Z[0] = B[0]
      U = np.exp(X@B)
      P = U/np.sum(U, axis=1).reshape(-1,1)
      B = (B + (a*((X.T)@(Y-P) - 2*l*(B-Z))))
    C = B
    # C = np.delete(C, 0, 0)
    Blist.append(C)
  return([np.array(Blist), P, Z])
```

```python
lam = [.0001, .001, .01, .1, 1, 10, 100, 1000, 10000]

v = gradD(df_train, df_train, lam)[0].T
print(np.shape(v))
```

```python
# Design Matrix

def dmat_prep2(df_fit, df_test):
  if isinstance(df_fit, pd.DataFrame):
    dmat_fit = df_fit.drop('Ancestry', axis = 1)
    dmat_test = df_test.drop('Ancestry', axis = 1)
  else:
    dmat_fit = np.delete(df_fit, 10, 1)
    dmat_fit = pd.DataFrame(dmat_fit, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
    dmat_fit = dmat_fit.drop('Ancestry', axis = 1)

    dmat_test = np.delete(df_test, 10, 1)
    dmat_test = pd.DataFrame(dmat_test, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
    dmat_test = dmat_test.drop('Ancestry', axis = 1)

  if isinstance(df_fit, np.ndarray):
    r_fit = pd.DataFrame(df_fit, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
  else:
    r_fit = df_fit
  rmat_fit = []
  for x in r_fit['Ancestry']:
    if x == 'African':
      rmat_fit.append([1, 0, 0, 0, 0])
    if x == 'European':
      rmat_fit.append([0, 1, 0, 0, 0])
    if x == 'EastAsian':
      rmat_fit.append([0, 0, 1, 0, 0])
    if x == 'Oceanian':
      rmat_fit.append([0, 0, 0, 1, 0])
    if x == 'NativeAmerican':
      rmat_fit.append([0, 0, 0, 0, 1])

  if isinstance(df_test, np.ndarray):
    r_test = pd.DataFrame(df_test, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10', 'Ancestry'])
  else:
    r_test = df_test
  rmat_test = []
  for x in r_test['Ancestry']:
    if x == 'African':
      rmat_test.append([1, 0, 0, 0, 0])
    if x == 'European':
      rmat_test.append([0, 1, 0, 0, 0])
    if x == 'EastAsian':
      rmat_test.append([0, 0, 1, 0, 0])
    if x == 'Oceanian':
      rmat_test.append([0, 0, 0, 1, 0])
    if x == 'NativeAmerican':
      rmat_test.append([0, 0, 0, 0, 1])


  
  for i in ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10']:
    dmat_test[i] = dmat_test[i].transform(lambda x: (x-dmat_fit[i].mean())/dmat_fit[i].std())
    dmat_fit[i] = dmat_fit[i].transform(lambda x: (x-x.mean())/x.std())

  dmat_fit = dmat_fit.to_numpy()
  dmat_test = dmat_test.to_numpy()
  ones_fit = np.ones(np.shape(df_fit)[0])
  ones_test = np.ones(np.shape(df_test)[0])
  dmat_fit = np.hstack((ones_fit[:, np.newaxis], dmat_fit))
  dmat_test = np.hstack((ones_test[:, np.newaxis], dmat_test))

  return([dmat_fit, rmat_fit, dmat_test, rmat_test])
```

```python
def fold_split(df):
  df1 = df.sample(frac = 1)

  # Split the folds
  f1 =dmat_prep2(df1.iloc[37:183], df1.iloc[:37])
  f2 =dmat_prep2(pd.concat([df1.iloc[:37], df1.iloc[74:183]]), df1.iloc[37:74])
  f3 =dmat_prep2(pd.concat([df1.iloc[:74], df1.iloc[111:183]]), df1.iloc[74:111])
  f4 =dmat_prep2(pd.concat([df1.iloc[:111], df1.iloc[147:183]]), df1.iloc[111:183])
  f5 =dmat_prep2(df1.iloc[:147], df1.iloc[147:183])

  # List of folds
  flist = [f1, f2, f3, f4, f5] 

  return(flist)
```

```python
training_folds = fold_split(df_train)

CVlist = []
for l in lam:
  g = 0
  for x in training_folds:
    b = gradD(x[0], x[1], [l])
    b = b[0][0]
    U = np.exp(x[2]@b)
    P = U/np.sum(U, axis=1).reshape(-1,1)
    CCE = np.sum(x[3]*np.log10(P))*(-1/np.shape(x[3])[0])
  CV = np.array(CCE)

  CVlist.append(CV)
```
